## Computation Graph Notebook

This notebook provides a detailed exploration of computation graphs used in deep learning. Below, you will find functions defined with appropriate documentation, visualizations of the computation graph, and interactive widgets to manipulate input values.


A computation graph is a visual representation of mathematical operations in a computational model, where nodes represent operations and edges represent data flow between these operations. It is commonly used in deep learning frameworks to optimize and execute complex mathematical expressions efficiently.

A computation graph is helpful for backward propagation in deep learning because it allows for efficient calculation of gradients using the chain rule by propagating gradients backwards through the graph from the output to the input nodes. This process simplifies the computation of gradients for complex mathematical expressions, making it easier to update model parameters during training.


```python
def J(a,b,c):
    return 3 * (a + b*c)

J(5,3,2)
```

```python
#The function J can be split into three seperate computaions:

# Function to compute the product of b and c
def U(b, c):
    return b * c

# Function to compute the sum of a and the product of b and c
def V(a, U):
    return a + U

# Function to compute three times the sum of a and the product of b and c
def J(V):
    return 3 * V

u = U(3, 2)
v = V(5, u)
j = J(v)
print(j)
```

```python
from graphviz import Digraph

dot = Digraph(comment='The Computation Graph')
dot.attr(rankdir='LR')  # Set graph orientation from Left to Right

dot.node('A', 'a')
dot.node('B', 'b')
dot.node('C', 'c')
dot.node('U', 'U(b, c)')
dot.node('V', 'V(a, U)')
dot.node('J', 'J(3, V)')

dot.edges(['BU', 'CU', 'AV', 'UV', 'VJ'])

# Display the graph within the Jupyter Notebook
dot
```

chain rule - a fundamental concept in calculus that allows you to find the derivative of a composite function. It states that the derivative of a composite function is the product of the derivatives of the outer and inner functions. In the context of neural networks and deep learning, the chain rule is crucial for calculating gradients efficiently during backpropagation.


```python
#dj/dv derivative 
dv = 3

#dj/da derivative 
dv_da = 1
da = dv * dv_da

#dj/du derivative 
du_dv = 1
du = dv * du_dv

#dj/db derivative 
du_db = 2
db = du * du_db

#dj/dc derivative 
du_dc = 3
dc = du * du_dc

print(da, db, dc)
```
