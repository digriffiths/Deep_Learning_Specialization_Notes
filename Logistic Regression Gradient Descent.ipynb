## Logistic Regression Gradient Descent


```python
from graphviz import Digraph

dot = Digraph(comment='The Computation Graph')
dot.attr(rankdir='LR')  # Set graph orientation from Left to Right

dot.node('x1', 'x1')
dot.node('w1', 'w1')
dot.node('x2', 'x2')
dot.node('w2', 'w2')
dot.node('b', 'b')

dot.node('Z', 'z = x1*w1 + x2*w2 + b')
dot.node('a', 'a = sigmoid(z)')
dot.node('Loss', 'L(a,y) = - (y * np.log(a) + (1 - y) * np.log(1 - a))')

# Connect all nodes to 'Z' correctly
dot.edge('x1', 'Z')
dot.edge('w1', 'Z')
dot.edge('x2', 'Z')
dot.edge('w2', 'Z')
dot.edge('b', 'Z')
dot.edge('Z', 'a')
dot.edge('a', 'Loss')

# Display the graph within the Jupyter Notebook
dot
```
