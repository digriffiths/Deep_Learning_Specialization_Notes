# Manual Calculation of Linear Regression for House Pricing Data

This Jupyter notebook demonstrates the process of performing simple linear regression manually using Python. It starts by generating synthetic data representing house sizes and their corresponding prices. The notebook then proceeds to calculate the mean of the house sizes and prices, which are used to determine the slope and intercept of the line of best fit according to the least squares method. Finally, it plots the original data points along with the manually calculated line of best fit to visually assess the accuracy of the model.


```python
import matplotlib.pyplot as plt
import numpy as np

house_sizes = np.array([750, 800, 900, 1050, 1150])
house_prices = np.array([200000, 220000, 230000, 270000, 320000])
```

The line of best fit will pass through the mean of the x-values (mean_x) and the mean of the y-values (mean_y)


```python
mean_x = house_sizes.mean()
mean_y = house_prices.mean()
```

The slope of the line of best fit is calculated using the formula:

slope = sum((xi - mean_x) \* (yi - mean_y)) / sum((xi - mean_x)^2)


```python
numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(house_sizes, house_prices))
denominator = sum((xi - mean_x) ** 2 for xi in house_sizes)
slope = numerator / denominator
```

The intercept is calculated


```python
intercept = mean_y - slope * mean_x
```

Calculate two points to plot the line of best fit using the minimum and maximum house sizes.


```python
x_fit = (min(house_sizes), max(house_sizes))
y_fit = [x * slope + intercept for x in x_fit]
```

Plot the original data points and the calculated line of best fit.


```python
plt.figure(figsize=(10, 5))
plt.scatter(house_sizes, house_prices, marker='o')

# Plotting the line of best fit
plt.plot(x_fit, y_fit, label='Manual Line of Best Fit', color='red')

# Adding labels and title
plt.title('House Prices vs. Size of House')
plt.xlabel('Size of House (sq ft)')
plt.ylabel('House Price ($)')
plt.grid(True)
plt.legend()  # Show legend
plt.show()
```
