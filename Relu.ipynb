# Relu

Rectified Linear Unit


ReLU is a type of activation function used in machine learning, particularly in neural networks, to introduce non-linearity. It allows the model to learn complex patterns in the data. The function outputs zero for any negative input and outputs the input itself for any non-negative input, which can be seen as setting a minimum threshold (zero in this case) for activation. It only allows positive signals to pass through, effectively setting a zero threshold for activation.


```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Generate some values
x = np.linspace(-10, 10, 100)
y = relu(x)

# Plotting
plt.plot(x, y)
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)
plt.show()
```
